```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Example dataset
data = {
    "Country": ["USA", "Canada", "India", "USA", "Canada", "India"],
    "Gender": ["Male", "Female", "Female", "Male", "Female", "Male"]
}
```

```python
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
encoder = OneHotEncoder()
encoded_array = encoder.fit_transform(df[["Country", "Gender"]]).toarray()
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(["Country", "Gender"]))
final_df = pd.concat([df, encoded_df], axis=1)
print(final_df)
```
