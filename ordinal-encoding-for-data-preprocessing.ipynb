```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Example dataset
data = {
    "Education": ["High School", "Bachelor", "Master", "PhD", "Bachelor", "Master"],
    "Satisfaction": ["Poor", "Average", "Good", "Excellent", "Good", "Average"]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# Define the order for each column
education_order = [["High School", "Bachelor", "Master", "PhD"]]
satisfaction_order = [["Poor", "Average", "Good", "Excellent"]]
```

```python
# Define the order for each column
education_order = [["High School", "Bachelor", "Master", "PhD"]]
satisfaction_order = [["Poor", "Average", "Good", "Excellent"]]
```

```python
# Initialize OrdinalEncoder with category orders
encoder = OrdinalEncoder(categories=education_order + satisfaction_order)

# Transform the dataset
df_encoded = encoder.fit_transform(df)

# Put results back into DataFrame
df[["Education_encoded", "Satisfaction_encoded"]] = df_encoded

print("\nDataFrame after Ordinal Encoding:")
print(df)
```
