```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/study-and-hours/study_hours_large.csv")
df
```

```python
X = df.Hours_Studied
y = df.Exam_Score
```

```python
m = 0
c = 0
Learning_rate = 0.01
iteration = 1000
n = int(len(X))
```

```python
for i in range(iteration):
    prediction_y = m * X + c
    derivation_m = (-2/n) * sum(X * (y - prediction_y))
    derivation_c = (-2/n) * sum(y - prediction_y)
    m = m - Learning_rate * derivation_m
    c = c - Learning_rate * derivation_c
```

```python
x = 9
y = m * x + c
print(y)
```
